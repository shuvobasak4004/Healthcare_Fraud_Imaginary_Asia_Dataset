```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# Number of rows to generate
num_rows = 20

# Generate random data for each field
data = {
    'Provider_ID': [f'PROV_{i}' for i in range(1, num_rows + 1)],
    'Claim_ID': [f'CLAIM_{i}' for i in range(1, num_rows + 1)],
    'Patient_ID': [f'PATIENT_{i}' for i in range(1, num_rows + 1)],
    'Diagnosis_Code': [f'DX_{random.randint(100, 999)}' for _ in range(num_rows)],
    'Procedure_Code': [f'PROC_{random.randint(1000, 9999)}' for _ in range(num_rows)],
    'Claim_Date': [datetime.now().strftime('%Y-%m-%d') for _ in range(num_rows)],
    'Admission_Date': [(datetime.now() - timedelta(days=random.randint(1, 30))).strftime('%Y-%m-%d') for _ in range(num_rows)],
    'Discharge_Date': [(datetime.now() + timedelta(days=random.randint(1, 30))).strftime('%Y-%m-%d') for _ in range(num_rows)],
    'Claim_Amount': [round(random.uniform(100, 10000), 2) for _ in range(num_rows)],
    'Paid_Amount': [round(random.uniform(50, 10000), 2) for _ in range(num_rows)],
    'Provider_Specialty': [random.choice(['Cardiology', 'Orthopedics', 'General Medicine']) for _ in range(num_rows)],
    'Patient_Age': [random.randint(18, 90) for _ in range(num_rows)],
    'Patient_Gender': [random.choice(['Male', 'Female']) for _ in range(num_rows)],
    'Fraud_Label': [random.choice([0, 1]) for _ in range(num_rows)],
    'Investigation_Details': [random.choice(['Under investigation', 'Cleared', 'Suspicious']) for _ in range(num_rows)],
    'Policy_Type': [random.choice(['HMO', 'PPO']) for _ in range(num_rows)],
    'Coverage_Amount': [round(random.uniform(1000, 5000), 2) for _ in range(num_rows)],
    'Total_Charges': [round(random.uniform(100, 10000), 2) for _ in range(num_rows)],
    'Payment_Type': [random.choice(['Credit Card', 'Check', 'Electronic Funds Transfer']) for _ in range(num_rows)],
    'State': [random.choice(['NY', 'CA', 'TX', 'FL', 'IL']) for _ in range(num_rows)]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Store the DataFrame in a CSV file
df.to_csv('healthcare_fraud_data.csv', index=False)

# Display the first few rows of the DataFrame
df.head()
```
